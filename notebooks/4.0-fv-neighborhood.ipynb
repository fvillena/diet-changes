```python
import numpy as np
import pandas as pd
import scipy.spatial.distance
```

```python
data = pd.read_csv(r'../data/processed/dataset_2d_reduction-pca_scaler-minmax_continent-None_cut-0.05-0.05_zanan-False.csv')
```

```python
continents = {}
for index,row in data.iterrows():
    continents[row['Area']] = row['Continent']
    
years = list(set(data['Year Code'].values))
metrics_total = [] 
for year in years:
    positions = np.array(data[data['Year Code'] == year][['x','y']])
    countries = data[data['Year Code'] == year].Area.values
    distances = scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(positions))
    lookupTable, indexed_dataSet = np.unique(countries, return_inverse=True)
    neighbors = pd.DataFrame(list(zip(
        countries,
        lookupTable[np.argsort(distances,axis=0)[1]],
        lookupTable[np.argsort(distances,axis=0)[2]],
        lookupTable[np.argsort(distances,axis=0)[3]],
        lookupTable[np.argsort(distances,axis=0)[4]],
        lookupTable[np.argsort(distances,axis=0)[5]],
    )),columns=['country']+list(range(0,5)))
    for i in range(5):
        neighbors[i] = neighbors[i].map(continents)
    neighbors['continent'] = neighbors.country.map(continents)
    for index,row in neighbors.iterrows():
        current_neighbors = [row[i] for i in range(5)]
        metric = current_neighbors.count(row['continent'])/5
        metrics_total.append((row['country'],year,metric))
metrics_total = pd.DataFrame(metrics_total,columns=['country','year','metric'])
metrics_total = pd.pivot_table(metrics_total,
     values='metric',
     index='country',
     columns='year').reset_index()
```
