```python
import pandas as pd
import numpy as np
import scipy.spatial
```

```python
data = pd.read_csv(r'../data/processed/dataset_2d_reduction-pca_scaler-minmax_continent-None_cut-0.05-0.05_zanan-False.csv')
```

```python
def euclideanDistance(a,b):
    return np.linalg.norm(a-b)
```

```python
def distance(v1,v2,metric):
    v1 = np.array(v1)
    v2 = np.array(v2)
    return metric(v1,v2)
```

```python
countries = list(set(data.Area.values))
```

```python
movements = []
for country in countries:
    distances = []
    positions = []
    for index,row in data[data.Area == country].iterrows():
        positions.append((row['x'],row['y']))
    for i in range(1,len(positions)):
        d = distance(positions[i-1], positions[i], euclideanDistance)
        distances.append(d)
    movement = sum(distances)
    movements.append((country,movement))
movements = sorted(movements, key=lambda x: x[1], reverse=True)
```
